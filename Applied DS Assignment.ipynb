### CORRELATION HEATMAP

```python
def correlation_heatmap(df):
    plt.figure(figsize=(12, 9))
    sns.heatmap(df.corr(), annot=True, cmap='viridis', annot_kws={"size": 14}, linewidths=0.5) 
    plt.xticks(fontsize=14, fontweight='bold') 
    plt.yticks(fontsize=14, fontweight='bold')  
    plt.title('Correlation Heatmap', fontsize=18, fontweight='bold') 
    plt.tight_layout()  
    plt.show()
```

```python
correlation_heatmap(df_test)
```

```python
# Categorizing age 

def age_category(age):
    if age <= 30:
        return "Young"
    elif age <= 60:
        return "Middle_aged"
    else:
        return "Old"
```

```python
def agewise_category(df):
    df["AGE_GROUP"] = df["AGE"].apply(age_category)
    return df.head()
```

```python
agewise_category(df)
```

```python
# Moving the column position of AGE_GROUP

column_to_move = df.pop("AGE_GROUP")

df.insert(2, "AGE_GROUP", column_to_move)
```

```python
# Changing the numerical values to categorical values for better understanding 

def val_change(df):
    mapping = {1: 'NO', 2: 'YES'}
    columns = ['SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE', 'CHRONIC DISEASE', 
               'FATIGUE ', 'ALLERGY ', 'WHEEZING', 'ALCOHOL CONSUMING', 'COUGHING', 
               'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN']
    
    columns_in_df = [col for col in columns if col in df.columns]
    df[columns_in_df] = df[columns_in_df].replace(mapping)
    return df
```

```python
val_change(df)
```

```python
# Bar plot to display the Gender difference

def gender_diff(df):
    fig, ax = plt.subplots(figsize=(5, 5))
    sns.barplot(x=df['GENDER'].value_counts().keys(), 
                y=df['GENDER'].value_counts(), 
                palette='Set2', 
                ax=ax, 
                width=0.5)
    ax.set_title("GENDER DIFFERENCE")
    ax.set_ylabel('COUNT')
    plt.tight_layout()
    plt.show()
```

```python
gender_diff(df)
```

```python
# Bar plot to visualize the Lung Cancer possibility

def lung_cancer_possibility(df):
    fig, ax = plt.subplots(figsize=(5, 5))
    sns.barplot(x=df['LUNG_CANCER'].value_counts().keys(), y=df['LUNG_CANCER'].value_counts(), palette = "hls", ax = ax, width = 0.5)
    plt.title("LUNG CANCER POSSIBILITY")
    plt.tight_layout()
    plt.ylabel('COUNT')
    plt.show()
```

```python
# Visualization for Lung Cancer Counts by Age Group

def age_based_lung_cancer(df):
    count_lung_cancer = pd.crosstab(df['AGE_GROUP'], df['LUNG_CANCER'])
    
    count_lung_cancer.plot(kind='bar', stacked=True, color=['green', 'red'], figsize=(6, 4))
    
    plt.title('Lung Cancer Counts by Age Group', fontsize=16)
    plt.xlabel('Age Group', fontsize=14)
    plt.ylabel('Count', fontsize=14)
    plt.xticks(rotation=0)
    plt.legend(title='Lung Cancer', labels=['No', 'Yes'])
    plt.grid(axis='y', linestyle='-', alpha=0.5)
    
    plt.tight_layout()
    plt.show()
```

```python
age_based_lung_cancer(df_cat)
```

```python
# Stacked Bar plot for visualizing the symptoms causing lung cancer

def symptom_lung_cancer(df):
    symptoms = ['YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE', 'CHRONIC DISEASE', 
            'FATIGUE ', 'ALLERGY ', 'WHEEZING', 'ALCOHOL CONSUMING', 'COUGHING', 
            'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN']
    plt.figure(figsize=(24, 16)) 
    
    for i, symptom in enumerate(symptoms, 1):
        plt.subplot(4, 5, i)
        sns.countplot(data=df, x=symptom, hue='LUNG_CANCER', palette='Spectral')
        plt.title(f'Lung Cancer by {symptom}', fontsize=16, fontweight='bold') 
        plt.xlabel(symptom, fontsize=14, fontweight = 'bold')  
        plt.ylabel('Count', fontsize=14, fontweight = 'bold')  
        plt.xticks(fontsize=14, fontweight = 'bold')  
        plt.yticks(fontsize=14, fontweight = 'bold')  
        plt.legend(title='Lung Cancer', labels=['No', 'Yes'], fontsize=14, title_fontsize=16) 
    
    plt.tight_layout()
    plt.show()
```

```python
symptom_lung_cancer(df_cat)
```

```python
# Histogram for the Age Distribution 

def age_dist(df):
    plt.figure(figsize=(10, 6))
    plt.hist(df['AGE'], bins=20, color='skyblue', edgecolor='black', density=True, alpha=0.6)
    sns.kdeplot(df['AGE'], color='red', linewidth=2)
    plt.title('Age Distribution', fontsize=18, fontweight='bold')
    plt.xlabel('Age', fontsize=14, fontweight='bold')
    plt.ylabel('Density', fontsize=14, fontweight='bold')
    plt.grid(axis='y', alpha=0.75)
    plt.show()
```

```python
age_dist(df_num)
```

```python
# Grouping the AGE column and the Lung cancer column
# Line plot for the Lung cancer caused by the Age

df_plot = pd.read_csv('lung cancer data.csv')

def age_based_plot(df):
    age_groups = df.groupby('AGE')['LUNG_CANCER'].value_counts().unstack()
    plt.figure(figsize=(6, 4))
    plt.plot(age_groups.index, age_groups['YES'], label='Lung Cancer Cases', color='red', marker='o')
    plt.plot(age_groups.index, age_groups['NO'], label='No Lung Cancer', color='blue', marker='o')
    plt.title('Lung Cancer Cases by Age', fontsize=16)
    plt.xlabel('Age', fontsize=14)
    plt.ylabel('Number of Cases', fontsize=14)
    plt.legend()
    plt.grid(visible=True, alpha=0.3)
    plt.show()
```

```python
age_based_plot(df_plot)
```
